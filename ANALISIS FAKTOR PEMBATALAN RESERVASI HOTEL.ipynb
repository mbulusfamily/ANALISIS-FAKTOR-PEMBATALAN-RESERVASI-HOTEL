# 🔬 **BAB III. ANALISIS DATA DAN PEMBAHASAN**

## A. Persiapan dan Pembersihan Data

### A.1. Data Loading dan Overview

Tahap pertama dalam analisis adalah melakukan loading dataset dan eksplorasi awal untuk memahami struktur data, kualitas data, dan karakteristik dataset yang akan dianalisis.

```python
# A.1. DATA LOADING DAN OVERVIEW

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import os
from datetime import datetime

# Set display options
pd.set_option('display.max_columns', None)
plt.style.use('default')
sns.set_palette("husl")

print("📚 Libraries imported successfully!")

# Load dataset
zip_path = r'c:\Users\ASUS\Downloads\hotel_bookings.csv.zip'

try:
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        csv_files = [f for f in zip_ref.namelist() if f.endswith('.csv')]
        if csv_files:
            with zip_ref.open(csv_files[0]) as csv_file:
                df = pd.read_csv(csv_file)
            print(f"✅ Dataset loaded: {df.shape[0]:,} rows, {df.shape[1]} columns")
        else:
            print("❌ CSV file not found in zip archive")
except FileNotFoundError:
    print("❌ File not found. Please check the file path.")
except Exception as e:
    print(f"❌ Error loading data: {e}")

# Dataset overview
if 'df' in locals():
    print("\n📊 DATASET INFORMATION")
    print("=" * 50)
    print(f"Shape: {df.shape}")
    print(f"Memory: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
    
    print(f"\n📋 DATA TYPES SUMMARY")
    dtype_summary = df.dtypes.value_counts()
    for dtype, count in dtype_summary.items():
        print(f"  {dtype}: {count} columns")
    
    print(f"\n📝 SAMPLE DATA (First 3 rows)")
    display(df.head(3))
else:
    print("⚠️ Dataset not loaded.")
```

### A.2. Data Cleaning dan Preprocessing

Tahap ini melakukan pembersihan data untuk memastikan kualitas data yang optimal sebelum dilakukan analisis lebih lanjut. Proses meliputi deteksi missing values, penanganan duplikasi, dan data validation.

```python
# A.2. MISSING VALUES ANALYSIS & DATA CLEANING

if 'df' in locals():
    print("🔍 MISSING VALUES ANALYSIS")
    print("=" * 40)
    
    # Missing values analysis
    missing_data = df.isnull().sum()
    missing_pct = (missing_data / len(df)) * 100
    
    missing_df = pd.DataFrame({
        'Column': missing_data.index,
        'Missing_Count': missing_data.values,
        'Missing_Percent': missing_pct.values
    }).query('Missing_Count > 0').sort_values('Missing_Percent', ascending=False)
    
    if len(missing_df) > 0:
        print("⚠️ COLUMNS WITH MISSING VALUES:")
        for _, row in missing_df.iterrows():
            print(f"  {row['Column']}: {row['Missing_Count']:,} ({row['Missing_Percent']:.1f}%)")
        
        # Visualization
        if len(missing_df) > 0:
            plt.figure(figsize=(12, 5))
            plt.subplot(1, 2, 1)
            missing_df.plot(x='Column', y='Missing_Percent', kind='bar', color='salmon')
            plt.title('Missing Values by Column')
            plt.ylabel('Percentage (%)')
            plt.xticks(rotation=45)
            
            plt.subplot(1, 2, 2)
            plt.text(0.1, 0.9, f"MISSING VALUES SUMMARY\n{'='*25}\n" +
                     f"Total Columns: {len(df.columns)}\n" +
                     f"Columns with Missing: {len(missing_df)}\n" +
                     f"Complete Columns: {len(df.columns) - len(missing_df)}", 
                     transform=plt.gca().transAxes, fontsize=11, verticalalignment='top')
            plt.axis('off')
            plt.tight_layout()
            plt.show()
    else:
        print("✅ No missing values found!")
    
    # Data cleaning process
    print(f"\n🧹 DATA CLEANING PROCESS")
    print("=" * 30)
    
    # Create clean copy
    df_clean = df.copy()
    original_shape = df_clean.shape
    
    # 1. Remove duplicates
    duplicates = df_clean.duplicated().sum()
    if duplicates > 0:
        df_clean = df_clean.drop_duplicates()
        print(f"✅ Removed {duplicates:,} duplicate records")
    
    # 2. Handle missing values
    # Drop columns with >80% missing
    high_missing = df_clean.isnull().mean() > 0.8
    cols_to_drop = df_clean.columns[high_missing].tolist()
    if cols_to_drop:
        df_clean = df_clean.drop(columns=cols_to_drop)
        print(f"✅ Dropped {len(cols_to_drop)} high-missing columns")
    
    # Fill remaining missing values
    for col in df_clean.columns:
        if df_clean[col].isnull().sum() > 0:
            if df_clean[col].dtype in ['int64', 'float64']:
                df_clean[col].fillna(df_clean[col].median(), inplace=True)
            else:
                df_clean[col].fillna(df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown', inplace=True)
    
    # 3. Final summary
    print(f"\n📊 CLEANING SUMMARY")
    print(f"Original: {original_shape[0]:,} rows × {original_shape[1]} cols")
    print(f"Cleaned:  {df_clean.shape[0]:,} rows × {df_clean.shape[1]} cols")
    print(f"Missing values remaining: {df_clean.isnull().sum().sum()}")
    
    if df_clean.isnull().sum().sum() == 0:
        print("✅ Dataset is now clean and ready for analysis!")
    
else:
    print("⚠️ Dataset not loaded. Run previous cell first.")
```

## B. Analisis Deskriptif

### B.1. Tingkat Pembatalan Keseluruhan

Analisis tingkat pembatalan keseluruhan untuk memahami karakteristik dasar dataset dan mengidentifikasi pola pembatalan secara umum.

```python
# B.1. ANALISIS TINGKAT PEMBATALAN KESELURUHAN

if 'df_clean' in locals():
    print("🎯 CANCELLATION OVERVIEW ANALYSIS")
    print("=" * 50)
    
    # Overall cancellation statistics
    total_bookings = len(df_clean)
    total_cancellations = df_clean['is_canceled'].sum()
    cancel_rate = (total_cancellations / total_bookings) * 100
    
    print(f"📊 OVERALL STATISTICS:")
    print(f"  • Total Bookings: {total_bookings:,}")
    print(f"  • Total Cancellations: {total_cancellations:,}")
    print(f"  • Overall Cancellation Rate: {cancel_rate:.2f}%")
    print(f"  • Completed Bookings: {total_bookings - total_cancellations:,}")
    
    # Create summary visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Pie chart for overall cancellation
    labels = ['Not Canceled', 'Canceled']
    sizes = [total_bookings - total_cancellations, total_cancellations]
    colors = ['lightgreen', 'salmon']
    explode = (0, 0.1)  # explode the canceled slice
    
    ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.set_title('Overall Booking Status Distribution')
    
    # Bar chart for counts
    ax2.bar(labels, sizes, color=colors, alpha=0.7)
    ax2.set_title('Booking Counts by Status')
    ax2.set_ylabel('Number of Bookings')
    
    # Add value labels on bars
    for i, v in enumerate(sizes):
        ax2.text(i, v + max(sizes) * 0.01, f'{v:,}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    # Additional insights
    print(f"\n💡 KEY INSIGHTS:")
    print(f"  • Approximately {100-cancel_rate:.1f}% of bookings are completed successfully")
    print(f"  • Every {100/cancel_rate:.0f} bookings, {cancel_rate/100*100/cancel_rate:.0f} gets canceled")
    print(f"  • Revenue impact: Potential loss from {total_cancellations:,} cancellations")
    
else:
    print("⚠️ Clean dataset not available. Run data cleaning section first.")
```

### B.2. Perbandingan City Hotel vs Resort Hotel

Analisis komparatif antara City Hotel dan Resort Hotel untuk mengidentifikasi perbedaan pola pembatalan dan karakteristik operasional kedua jenis hotel.

```python
# B.2. PERBANDINGAN CITY HOTEL VS RESORT HOTEL

if 'df_clean' in locals() and 'hotel' in df_clean.columns:
    print("🏨 HOTEL TYPE COMPARISON ANALYSIS")
    print("=" * 50)
    
    # Calculate statistics by hotel type
    hotel_stats = df_clean.groupby('hotel')['is_canceled'].agg(['count', 'sum', 'mean']).round(3)
    hotel_stats['cancel_rate_pct'] = (hotel_stats['mean'] * 100).round(2)
    hotel_stats['completed_bookings'] = hotel_stats['count'] - hotel_stats['sum']
    
    print("📈 HOTEL TYPE STATISTICS:")
    print("-" * 70)
    print(f"{'Hotel Type':<15} {'Total':<10} {'Canceled':<10} {'Rate %':<8} {'Completed':<10}")
    print("-" * 70)
    
    for hotel in hotel_stats.index:
        stats = hotel_stats.loc[hotel]
        print(f"{hotel:<15} {stats['count']:<10,} {stats['sum']:<10,} {stats['cancel_rate_pct']:<8} {stats['completed_bookings']:<10,}")
    
    # Calculate difference
    if len(hotel_stats) == 2:
        rates = hotel_stats['cancel_rate_pct'].values
        difference = abs(rates[0] - rates[1])
        higher_hotel = hotel_stats.index[hotel_stats['cancel_rate_pct'].idxmax()]
        print(f"\n🔍 KEY DIFFERENCES:")
        print(f"  • {higher_hotel} has {difference:.2f}% higher cancellation rate")
        print(f"  • This represents a {difference/min(rates)*100:.1f}% relative difference")
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Cancellation rates by hotel type
    hotel_stats['cancel_rate_pct'].plot(kind='bar', ax=axes[0,0], color=['skyblue', 'lightcoral'])
    axes[0,0].set_title('Cancellation Rate by Hotel Type')
    axes[0,0].set_ylabel('Cancellation Rate (%)')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Volume comparison
    x = range(len(hotel_stats))
    width = 0.35
    axes[0,1].bar([i - width/2 for i in x], hotel_stats['count'], width, 
                  label='Total Bookings', alpha=0.8, color='steelblue')
    axes[0,1].bar([i + width/2 for i in x], hotel_stats['sum'], width, 
                  label='Cancellations', alpha=0.8, color='salmon')
    axes[0,1].set_title('Booking Volume vs Cancellations')
    axes[0,1].set_ylabel('Count')
    axes[0,1].set_xticks(x)
    axes[0,1].set_xticklabels(hotel_stats.index, rotation=45)
    axes[0,1].legend()
    
    # Average metrics comparison (if available)
    if 'adr' in df_clean.columns:
        adr_comparison = df_clean.groupby('hotel')['adr'].mean()
        adr_comparison.plot(kind='bar', ax=axes[1,0], color='gold')
        axes[1,0].set_title('Average Daily Rate by Hotel Type')
        axes[1,0].set_ylabel('ADR (€)')
        axes[1,0].tick_params(axis='x', rotation=45)
    
    if 'lead_time' in df_clean.columns:
        lead_comparison = df_clean.groupby('hotel')['lead_time'].mean()
        lead_comparison.plot(kind='bar', ax=axes[1,1], color='lightgreen')
        axes[1,1].set_title('Average Lead Time by Hotel Type')
        axes[1,1].set_ylabel('Lead Time (days)')
        axes[1,1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Additional analysis by hotel characteristics
    print(f"\n🏢 HOTEL CHARACTERISTICS:")
    for hotel in df_clean['hotel'].unique():
        hotel_data = df_clean[df_clean['hotel'] == hotel]
        print(f"\n{hotel.upper()}:")
        print(f"  • Market Focus: {'Business travelers' if hotel == 'City Hotel' else 'Leisure travelers'}")
        print(f"  • Total Bookings: {len(hotel_data):,}")
        print(f"  • Cancellation Rate: {hotel_data['is_canceled'].mean()*100:.2f}%")
        
        if 'lead_time' in hotel_data.columns:
            print(f"  • Avg Lead Time: {hotel_data['lead_time'].mean():.0f} days")
        if 'adr' in hotel_data.columns:
            print(f"  • Avg Daily Rate: €{hotel_data['adr'].mean():.0f}")
    
else:
    print("⚠️ Clean dataset not available or hotel column missing.")
```

## C. Analisis Faktor Pembatalan

### C.1. Analisis Lead Time

Analisis hubungan antara lead time (waktu antara booking dan kedatangan) dengan tingkat pembatalan reservasi.

```python
# C.1. ANALISIS LEAD TIME

if 'df_clean' in locals() and 'lead_time' in df_clean.columns:
    print("📅 LEAD TIME ANALYSIS")
    print("=" * 40)
    
    # Create lead time categories
    df_clean['lead_time_group'] = pd.cut(df_clean['lead_time'], 
                                       bins=[0, 7, 30, 90, 365, df_clean['lead_time'].max()],
                                       labels=['0-7 days', '8-30 days', '31-90 days', 
                                              '91-365 days', '>365 days'])
    
    # Lead time impact analysis
    lead_analysis = df_clean.groupby('lead_time_group').agg({
        'is_canceled': ['count', 'sum', 'mean'],
        'lead_time': 'mean'
    }).round(3)
    
    lead_analysis.columns = ['Total_Bookings', 'Cancellations', 'Cancel_Rate', 'Avg_Lead_Time']
    lead_analysis['Cancel_Rate_Pct'] = (lead_analysis['Cancel_Rate'] * 100).round(2)
    
    print("📊 CANCELLATION BY LEAD TIME GROUPS:")
    print("-" * 80)
    print(f"{'Lead Time Group':<15} {'Total':<8} {'Canceled':<10} {'Rate %':<8} {'Avg Days':<10}")
    print("-" * 80)
    
    for group in lead_analysis.index:
        stats = lead_analysis.loc[group]
        print(f"{group:<15} {stats['Total_Bookings']:<8,} {stats['Cancellations']:<10,} "
              f"{stats['Cancel_Rate_Pct']:<8} {stats['Avg_Lead_Time']:<10.0f}")
    
    # Statistical correlation
    correlation = df_clean['lead_time'].corr(df_clean['is_canceled'])
    print(f"\n🔗 LEAD TIME CORRELATION:")
    print(f"  • Correlation coefficient: {correlation:.3f}")
    print(f"  • Relationship: {'Positive' if correlation > 0 else 'Negative'}")
    print(f"  • Strength: {'Strong' if abs(correlation) > 0.5 else 'Moderate' if abs(correlation) > 0.3 else 'Weak'}")
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Cancellation rate by lead time group
    lead_analysis['Cancel_Rate_Pct'].plot(kind='bar', ax=axes[0,0], color='steelblue')
    axes[0,0].set_title('Cancellation Rate by Lead Time Group')
    axes[0,0].set_ylabel('Cancellation Rate (%)')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Volume distribution
    lead_analysis['Total_Bookings'].plot(kind='bar', ax=axes[0,1], color='lightgreen')
    axes[0,1].set_title('Booking Volume by Lead Time Group')
    axes[0,1].set_ylabel('Number of Bookings')
    axes[0,1].tick_params(axis='x', rotation=45)
    
    # Scatter plot: Lead time vs cancellation
    sample_data = df_clean.sample(n=min(5000, len(df_clean)))  # Sample for performance
    axes[1,0].scatter(sample_data['lead_time'], sample_data['is_canceled'], 
                     alpha=0.5, s=1, color='red')
    axes[1,0].set_xlabel('Lead Time (days)')
    axes[1,0].set_ylabel('Canceled (0/1)')
    axes[1,0].set_title('Lead Time vs Cancellation (Sample)')
    
    # Box plot by hotel type
    if 'hotel' in df_clean.columns:
        df_clean.boxplot(column='lead_time', by='hotel', ax=axes[1,1])
        axes[1,1].set_title('Lead Time Distribution by Hotel Type')
        axes[1,1].set_xlabel('Hotel Type')
        axes[1,1].set_ylabel('Lead Time (days)')
    
    plt.tight_layout()
    plt.show()
    
    # Key insights
    print(f"\n💡 KEY INSIGHTS:")
    highest_rate_group = lead_analysis['Cancel_Rate_Pct'].idxmax()
    lowest_rate_group = lead_analysis['Cancel_Rate_Pct'].idxmin()
    print(f"  • Highest cancellation rate: {highest_rate_group} ({lead_analysis.loc[highest_rate_group, 'Cancel_Rate_Pct']:.1f}%)")
    print(f"  • Lowest cancellation rate: {lowest_rate_group} ({lead_analysis.loc[lowest_rate_group, 'Cancel_Rate_Pct']:.1f}%)")
    print(f"  • Lead time shows {'positive' if correlation > 0 else 'negative'} correlation with cancellation")
    
else:
    print("⚠️ Clean dataset or lead_time column not available.")
```

### C.2. Analisis Average Daily Rate (ADR)

Analisis hubungan antara harga kamar (Average Daily Rate) dengan tingkat pembatalan untuk mengidentifikasi sensitivity harga terhadap perilaku pembatalan.

```python
# C.2. ANALISIS AVERAGE DAILY RATE (ADR)

if 'df_clean' in locals() and 'adr' in df_clean.columns:
    print("💰 AVERAGE DAILY RATE ANALYSIS")
    print("=" * 45)
    
    # ADR statistics by cancellation status
    adr_stats = df_clean.groupby(['hotel', 'is_canceled'])['adr'].agg(['count', 'mean', 'median', 'std']).round(2)
    
    print("📊 ADR STATISTICS BY HOTEL AND CANCELLATION STATUS:")
    print("-" * 80)
    print(f"{'Hotel':<12} {'Status':<12} {'Count':<8} {'Mean':<8} {'Median':<8} {'Std':<8}")
    print("-" * 80)
    
    for hotel in df_clean['hotel'].unique():
        for status in [0, 1]:
            if (hotel, status) in adr_stats.index:
                stats = adr_stats.loc[(hotel, status)]
                status_text = 'Not Canceled' if status == 0 else 'Canceled'
                print(f"{hotel:<12} {status_text:<12} {stats['count']:<8,} {stats['mean']:<8} {stats['median']:<8} {stats['std']:<8}")
    
    # Calculate price sensitivity
    print(f"\n💡 PRICE SENSITIVITY ANALYSIS:")
    for hotel in df_clean['hotel'].unique():
        hotel_data = df_clean[df_clean['hotel'] == hotel]
        if len(hotel_data) > 0:
            adr_not_canceled = hotel_data[hotel_data['is_canceled'] == 0]['adr'].mean()
            adr_canceled = hotel_data[hotel_data['is_canceled'] == 1]['adr'].mean()
            price_diff = adr_canceled - adr_not_canceled
            price_diff_pct = (price_diff / adr_not_canceled) * 100 if adr_not_canceled > 0 else 0
            
            print(f"\n{hotel}:")
            print(f"  • Not Canceled ADR: €{adr_not_canceled:.2f}")
            print(f"  • Canceled ADR: €{adr_canceled:.2f}")
            print(f"  • Difference: €{price_diff:.2f} ({price_diff_pct:+.1f}%)")
            print(f"  • Price Sensitivity: {'High' if abs(price_diff_pct) > 10 else 'Moderate' if abs(price_diff_pct) > 5 else 'Low'}")
    
    # ADR correlation with cancellation
    adr_correlation = df_clean['adr'].corr(df_clean['is_canceled'])
    print(f"\n🔗 ADR-CANCELLATION CORRELATION:")
    print(f"  • Correlation coefficient: {adr_correlation:.3f}")
    print(f"  • Interpretation: {'Higher prices increase cancellation risk' if adr_correlation > 0 else 'Higher prices decrease cancellation risk'}")
    
    # Create ADR price bands for analysis
    df_clean['adr_band'] = pd.cut(df_clean['adr'], 
                                 bins=[0, 50, 100, 150, 200, df_clean['adr'].max()],
                                 labels=['€0-50', '€51-100', '€101-150', '€151-200', '€200+'])
    
    band_analysis = df_clean.groupby('adr_band')['is_canceled'].agg(['count', 'mean']).round(3)
    band_analysis['cancel_rate_pct'] = (band_analysis['mean'] * 100).round(2)
    
    print(f"\n📈 CANCELLATION BY PRICE BANDS:")
    print("-" * 50)
    print(f"{'Price Band':<12} {'Bookings':<10} {'Cancel Rate %':<15}")
    print("-" * 50)
    for band in band_analysis.index:
        stats = band_analysis.loc[band]
        print(f"{band:<12} {stats['count']:<10,} {stats['cancel_rate_pct']:<15}")
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # ADR comparison by cancellation status
    adr_comparison = df_clean.groupby(['hotel', 'is_canceled'])['adr'].mean().unstack()
    adr_comparison.plot(kind='bar', ax=axes[0,0], color=['lightgreen', 'salmon'])
    axes[0,0].set_title('Average Daily Rate: Canceled vs Not Canceled')
    axes[0,0].set_ylabel('ADR (€)')
    axes[0,0].legend(['Not Canceled', 'Canceled'])
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Cancellation rate by price bands
    band_analysis['cancel_rate_pct'].plot(kind='bar', ax=axes[0,1], color='orange')
    axes[0,1].set_title('Cancellation Rate by Price Bands')
    axes[0,1].set_ylabel('Cancellation Rate (%)')
    axes[0,1].tick_params(axis='x', rotation=45)
    
    # ADR distribution by cancellation status
    canceled_adr = df_clean[df_clean['is_canceled'] == 1]['adr']
    not_canceled_adr = df_clean[df_clean['is_canceled'] == 0]['adr']
    
    axes[1,0].hist([not_canceled_adr, canceled_adr], bins=30, alpha=0.7, 
                   label=['Not Canceled', 'Canceled'], color=['lightgreen', 'salmon'])
    axes[1,0].set_title('ADR Distribution by Cancellation Status')
    axes[1,0].set_xlabel('ADR (€)')
    axes[1,0].set_ylabel('Frequency')
    axes[1,0].legend()
    
    # Box plot: ADR by hotel and cancellation
    df_clean.boxplot(column='adr', by=['hotel', 'is_canceled'], ax=axes[1,1])
    axes[1,1].set_title('ADR Distribution by Hotel and Cancellation')
    axes[1,1].set_xlabel('Hotel Type, Cancellation Status')
    axes[1,1].set_ylabel('ADR (€)')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("⚠️ Clean dataset or ADR column not available.")
```

```python
# C.3. ANALISIS SPECIAL REQUESTS DAN KORELASI FAKTOR

if 'df_clean' in locals():
    print("🎯 SPECIAL REQUESTS & CORRELATION ANALYSIS")
    print("=" * 55)
    
    # Special requests analysis
    if 'total_of_special_requests' in df_clean.columns:
        special_analysis = df_clean.groupby('total_of_special_requests')['is_canceled'].agg(['count', 'mean']).round(3)
        special_analysis['cancel_rate_pct'] = (special_analysis['mean'] * 100).round(2)
        
        print("🎁 SPECIAL REQUESTS IMPACT:")
        print("-" * 45)
        print(f"{'Requests':<10} {'Bookings':<10} {'Cancel Rate %':<15}")
        print("-" * 45)
        for requests in special_analysis.index:
            stats = special_analysis.loc[requests]
            print(f"{requests:<10} {stats['count']:<10,} {stats['cancel_rate_pct']:<15}")
        
        # Special requests correlation
        special_corr = df_clean['total_of_special_requests'].corr(df_clean['is_canceled'])
        print(f"\n🔗 Special Requests Correlation: {special_corr:.3f}")
        print(f"   Interpretation: {'More requests = less cancellation' if special_corr < 0 else 'More requests = more cancellation'}")
    
    # Overall correlation analysis
    print(f"\n📊 CORRELATION MATRIX - KEY FACTORS:")
    print("=" * 50)
    
    # Select numeric columns for correlation
    numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
    key_factors = ['is_canceled', 'lead_time', 'adr', 'total_of_special_requests', 
                   'booking_changes', 'adults', 'children', 'babies']
    available_factors = [col for col in key_factors if col in numeric_cols]
    
    if len(available_factors) > 2:
        correlation_matrix = df_clean[available_factors].corr()
        cancellation_corr = correlation_matrix['is_canceled'].drop('is_canceled')
        cancellation_corr = cancellation_corr.reindex(cancellation_corr.abs().sort_values(ascending=False).index)
        
        print("🔗 CORRELATIONS WITH CANCELLATION:")
        print("-" * 40)
        print(f"{'Factor':<25} {'Correlation':<12} {'Strength':<12}")
        print("-" * 40)
        
        for factor, corr in cancellation_corr.items():
            strength = 'Strong' if abs(corr) > 0.5 else 'Moderate' if abs(corr) > 0.3 else 'Weak'
            direction = '+' if corr > 0 else '-'
            print(f"{factor:<25} {direction}{abs(corr):<11.3f} {strength:<12}")
    
    # Market segment analysis
    if 'market_segment' in df_clean.columns:
        print(f"\n🎯 MARKET SEGMENT ANALYSIS:")
        segment_rates = df_clean.groupby('market_segment')['is_canceled'].mean() * 100
        segment_rates = segment_rates.sort_values(ascending=False)
        
        print("-" * 35)
        print(f"{'Market Segment':<20} {'Cancel Rate %':<12}")
        print("-" * 35)
        for segment, rate in segment_rates.head(6).items():
            print(f"{segment:<20} {rate:<12.1f}")
    
    # Customer type analysis
    if 'customer_type' in df_clean.columns:
        print(f"\n👥 CUSTOMER TYPE ANALYSIS:")
        customer_rates = df_clean.groupby('customer_type')['is_canceled'].mean() * 100
        customer_rates = customer_rates.sort_values(ascending=False)
        
        print("-" * 30)
        print(f"{'Customer Type':<15} {'Cancel Rate %':<12}")
        print("-" * 30)
        for ctype, rate in customer_rates.items():
            print(f"{ctype:<15} {rate:<12.1f}")
    
    # Comprehensive visualization
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Special requests impact
    if 'total_of_special_requests' in df_clean.columns:
        special_analysis['cancel_rate_pct'].plot(kind='bar', ax=axes[0,0], color='purple')
        axes[0,0].set_title('Cancellation Rate by Special Requests')
        axes[0,0].set_ylabel('Cancellation Rate (%)')
        axes[0,0].set_xlabel('Number of Special Requests')
    
    # Correlation heatmap
    if len(available_factors) > 2:
        corr_subset = correlation_matrix.loc[available_factors[:6], available_factors[:6]]
        sns.heatmap(corr_subset, annot=True, cmap='RdBu_r', center=0,
                   square=True, linewidths=0.5, ax=axes[0,1], fmt='.2f')
        axes[0,1].set_title('Correlation Matrix - Key Factors')
    
    # Market segment cancellation rates
    if 'market_segment' in df_clean.columns:
        segment_rates.head(5).plot(kind='bar', ax=axes[1,0], color='orange')
        axes[1,0].set_title('Top 5 Market Segments by Cancellation')
        axes[1,0].set_ylabel('Cancellation Rate (%)')
        axes[1,0].tick_params(axis='x', rotation=45)
    
    # Customer type analysis
    if 'customer_type' in df_clean.columns:
        customer_rates.plot(kind='bar', ax=axes[1,1], color='teal')
        axes[1,1].set_title('Cancellation Rate by Customer Type')
        axes[1,1].set_ylabel('Cancellation Rate (%)')
        axes[1,1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Key insights summary
    print(f"\n💡 KEY INSIGHTS FROM FACTOR ANALYSIS:")
    if 'total_of_special_requests' in df_clean.columns:
        avg_special_not_canceled = df_clean[df_clean['is_canceled'] == 0]['total_of_special_requests'].mean()
        avg_special_canceled = df_clean[df_clean['is_canceled'] == 1]['total_of_special_requests'].mean()
        print(f"  • Special requests: Not canceled avg = {avg_special_not_canceled:.2f}, Canceled avg = {avg_special_canceled:.2f}")
    
    if len(available_factors) > 2:
        strongest_factor = cancellation_corr.abs().idxmax()
        strongest_corr = cancellation_corr[strongest_factor]
        print(f"  • Strongest predictor: {strongest_factor} (r = {strongest_corr:.3f})")
    
    print(f"  • Customer engagement (special requests) appears to reduce cancellation risk")
    print(f"  • Different market segments show significantly different cancellation patterns")
    
else:
    print("⚠️ Clean dataset not available.")
```
